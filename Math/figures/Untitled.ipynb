```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Variables
x, y, lam = sp.symbols('x y lam')

# Objective function
f = 1 - x**2 - y**2

# Constraint
g = x + y - 1

# Lagrange function
L = f - lam * g

# Calculate the partial derivatives
grad_L = [sp.diff(L, var) for var in (x, y, lam)]

# Solve the system of equations
solutions = sp.solve(grad_L)

# Print the solutions
#for sol in solutions:
#    print(f'x = {sol[x]}, y = {sol[y]}, f = {f.subs(sol)}')

# Prepare for plotting
x_, y_ = np.arange(-2,2.01,0.01), np.arange(-2,2.01,0.01)
x_np, y_np = np.meshgrid(x_, y_)
f_np = 1 - x_np**2 - y_np**2
g_np = x_np + y_np - 1

# Plot the original contour
contour = plt.contour(x_np,y_np,f_np,cmap="viridis")
plt.clabel(contour, inline=True, fontsize=8)

# Plot the constraint
constraint = plt.contour(x_np, y_np, g_np, levels=[0], colors='r')
plt.clabel(constraint, inline=True, fontsize=8)

# Add the solutions to the plot
plt.plot(solutions[x], solutions[y], 'ko')

# Add a circle with radius sqrt(2)/2
circle = plt.Circle((0, 0), np.sqrt(2)/2, color='r', fill=False)
plt.gca().add_artist(circle)

plt.colorbar(contour)
plt.savefig("./lagrange_3.png", format="png", dpi=600)
```
